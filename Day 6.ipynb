# Pandas and Mat

```python
import pandas as pd
filePath="Datafile/RegularSeasonCompactResults.csv"
def readCsvData(filePath):
    return pd.read_csv(filePath)
df=readCsvData(filePath)
print(df)
```

```python
#GroupBy
df.groupby("Wteam")["Wscore"].mean().head()
```
